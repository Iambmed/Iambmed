#Analyzing The data


**I-KDE Plot**

```python
plt.figure(figsize=(15,3))
sns.kdeplot(market_data['Sales'], color='b', label='Sales',shade=True, bw_method=25)
sns.kdeplot(market_data['Profit'], color='Black', label='Profit', shade=True, bw_method=25)
plt.xlim([-100,1000])
plt.legend()
plt.show()
```

**II-Pairplot (Based on Category)**

```python
sns.pairplot(market_data,hue='Category')
plt.show()
```

**III- Pairpolt (Based on Region)**

```python
sns.pairplot(market_data,hue='Region')
plt.show()
```

**IV-Correlation**

```python
market_data.corr()
```

**V- Heatmap of Correlation**

```python
sns.heatmap(market_data.corr(), annot=True)
plt.show()
```

1-Sales and Profit are moderately Correlated

2-Quantity and Profit are moderately correlated

3-Discount and Profit are negative correlated

**VI- Countplot of columns**

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,7));

sns.countplot(market_data['Category'],ax=axs[0][0])
sns.countplot(market_data['Segment'],ax=axs[0][1])
sns.countplot(market_data['Ship Mode'],ax=axs[1][0])
sns.countplot(market_data['Region'],ax=axs[1][1])
axs[0][0].set_title('Category',fontsize=20)
axs[0][1].set_title('Segment',fontsize=20)
axs[1][0].set_title('Ship Mode',fontsize=20)
axs[1][1].set_title('Region',fontsize=20)

plt.tight_layout()
plt.show()
```
