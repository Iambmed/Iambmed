# Data Visualization

**I- No of Matches Bar Plot**

```python
plt.figure(figsize = (18,10))
sns.countplot('season',data=matches_data,palette="dark")
plt.title("Number of Matches played in Every IPL season",fontsize=20,color="DarkBlue")
plt.xlabel("season",fontsize=15,c='b')
plt.ylabel('Matches',fontsize=15,c='b')
plt.show()
```

**The highest number of match played in IPL season was 2013,2012,2011**

**II- No of Win Matches per Team**

```python

plt.figure(figsize = (18,10))
sns.countplot(x='winner',data=matches_data, palette='summer')
plt.title("Numbers of matches won by team ",fontsize=20,color="DarkBlue")
plt.xticks(rotation=50)
plt.xlabel("Teams",fontsize=15,c='b')
plt.ylabel("No of wins",fontsize=15,c='b')
plt.show()
```

**The highest number of match won by Mumbai Indians**

**III- Win Prediction PieChart**


```python
matches_data['win_by']=np.where(matches_data['win_by_runs']>0,'Bat first','Bowl first')
```

```python

Win=matches_data.win_by.value_counts()
labels=np.array(Win.index)
sizes = Win.values
colors = ['#0095FF', '#Ff0068']
plt.figure(figsize = (5,5))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20, color="DarkBlue")
plt.axis('equal')
plt.show()
```

**Teams which Bowl first has higher chances of winning than the team which bat first**

**IV- Comparison of Batting and Bowling First BarPlot**

```python
plt.figure(figsize = (17,5))
sns.countplot('season',hue='win_by',data=matches_data,palette='hsv')
plt.title("Numbers of matches won by batting and bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

**V- Toss Result PieChart**

```python
Toss=matches_data.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['#0095FF', '#Ff0068']
plt.figure(figsize = (5,5))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss result',fontsize=20,color="DarkBlue")
plt.axis('equal')
plt.show()
```

**After winning toss more teams decide to do fielding first**

**VI- Win Matches By Toss Result BarPlot**


```python
plt.figure(figsize = (15,5))
sns.countplot('season',hue='toss_decision',data=matches_data,palette='afmhot')
plt.title("Numbers of matches won by Toss result ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

**In finals teams which decide to do fielding first win the matches more then the team which bat first**

```python
final_matches=matches_data.drop_duplicates(subset=['season'], keep='last')

final_matches[['season','winner']].reset_index(drop=True).sort_values('season')
```

**VII- Winning Final Percentage PieChart**

```python
match = final_matches.win_by.value_counts()
labels=np.array(Toss.index)
sizes = match.values
colors = ['#0095FF', '#Ff0068']
plt.figure(figsize = (5,5))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20,color="DarkBlue")
plt.axis('equal')
plt.show()
```

**In finals most teams after winning toss decide to do fielding first**

**VIII- BarPlot of TopPlayers**


```python
plt.figure(figsize = (18,10))
top_players = matches_data.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count",c='b')
ax.set_title("Top player of the match Winners",color="DarkBlue")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="hsv");
plt.show()
```

**Top player of match winning are CH gayle, AB de villers**

```python
four_data=complete_data[complete_data['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])

```

**IX- Barplot of Number of 4 hit by Players**

```python
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color="m")
plt.title("Numbers of fours hit by playes ",fontsize=20,color="DarkBlue")
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15,c='b')
plt.ylabel("No of fours",fontsize=15,c='b')
plt.show()
```

**The highest number of four hit by player is Shikar Dhawan**

**X- Barplot of Number of 4 Hit by Each Player**

```python
ax=four_data.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',kind='bar',color = "m")
plt.title("Numbers of fours hit in each season ",fontsize=20,color="DarkBlue")
plt.xticks(rotation=50)
plt.xlabel("season",fontsize=15,c='b')
plt.ylabel("No of fours",fontsize=15,c='b')
plt.show()
```

```python
six_data=complete_data[complete_data['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

**XI- Barplot of Number of 6 hit by Each Player**

```python
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color="m")
plt.title("Numbers of six hit by playes ",fontsize=20, color="DarkBlue")
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15,c='b')
plt.ylabel("No of six",fontsize=15,c='b')
plt.show()
```

**The highest number of six hit by player is CH gayle**

**XII- BarPlot of Number of 6 hit by Each Season**

```python
ax=six_data.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind='bar',color = "m")
plt.title("Numbers of six hit in each season ",fontsize=20,color="DarkBlue")
plt.xticks(rotation=50)
plt.xlabel("season",fontsize=15,c='b')
plt.ylabel("No of six",fontsize=15,c='b')
plt.show()
```

```python
batsman_score=deliveries_data.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("*** Top 10 Leading Run Scorer in IPL ***")
batsman_score.iloc[:10,:]
```

**Top leading run scorer in IPL are Virat kholi, SK Raina, RG Sharma**

```python
No_Matches_player= deliveries_data[["match_id","player_dismissed"]]
No_Matches_player =No_Matches_player .groupby("player_dismissed")["match_id"].count().reset_index().sort_values(by="match_id",ascending=False).reset_index(drop=True)
No_Matches_player.columns=["batsman","No_of Matches"]
No_Matches_player .head(5)
```

**The highest number of matches played by player name are SK Raina, RG Sharma**

**XIII- BarPlot of Most Dismissals by IPL**

```python
plt.figure(figsize=(14,5))
ax=sns.countplot(deliveries_data.dismissal_kind)
plt.title("Dismissals in IPL",fontsize=20,color="DarkBlue")
plt.xlabel("Dismissals kind",fontsize=15,c='b')
plt.ylabel("count",fontsize=15,c='b')
plt.xticks(rotation=90)
plt.show()
```

**Dismissals in IPL was most by Catch out**

```python
wicket_data=deliveries_data.dropna(subset=['dismissal_kind'])
wicket_data=wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]

```

```python
wicket_data.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True).iloc[:10,:]
```

**The IPL most wicket taken blower is SL Malinga**
